```python
import pandas as pd

def load_data():
    url = "https://raw.githubusercontent.com/kelvins/municipios-brasileiros/refs/heads/main/csv/municipios.csv"
    municipios = pd.read_csv(url)
    url_estados = "https://raw.githubusercontent.com/kelvins/municipios-brasileiros/refs/heads/main/csv/estados.csv"
    estados = pd.read_csv(url_estados)
    populacao = pd.read_csv("https://raw.githubusercontent.com/mapaslivres/municipios-br/refs/heads/main/tabelas/municipios.csv")
    return municipios, estados, populacao

municipios, estados, populacao = load_data()
```

```python
# Merge dos dados
df_merged = municipios.merge(estados, on='codigo_uf', how='left')
df_merged = df_merged.rename(columns={
    'nome_x': 'municipio',
    'latitude_x': 'latitude',
    'longitude_x': 'longitude',
    'nome_y': 'estado',
    'latitude_y': 'latitude_estado',
    'longitude_y': 'longitude_estado'
})

df_merged
```

```python
# Em populacao, renomea a coluna municipio para codigo_ibge
populacao = populacao.rename(columns={'municipio': 'codigo_ibge'})
```

```python
# Merge com df_merged
df_merged = df_merged.merge(populacao, on='codigo_ibge', how='left')

df_merged
```

```python
# Remove as colunas uf_y, name, mesoregion, microregion, rgint, rgi, osm_relation_id, wikidata_id, is_capital, wikipedia_pt, lon, lat, no_accents, slug_name, alternative_names
df_merged = df_merged.drop(columns=['uf_y', 'name', 'mesoregion', 'microregion', 'rgint', 'rgi', 'osm_relation_id', 'wikidata_id', 'is_capital', 'wikipedia_pt', 'lon', 'lat', 'no_accents', 'slug_name', 'alternative_names'])

df_merged
```

```python
# Renomeia uf_x para uf
df_merged = df_merged.rename(columns={'uf_x': 'uf'})
df_merged
```

```python
# remove uf_code
df_merged = df_merged.drop(columns=['uf_code'])
df_merged
```

```python
# salva em csv
df_merged.to_csv("dados_municipais.csv", index=False)
```
